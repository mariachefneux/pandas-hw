```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
total_players = purchase_data['SN'].nunique()

pd.DataFrame({
    'Total Players': [total_players]
})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
pd.DataFrame({
    'Number of Unique Items': [purchase_data['Item ID'].nunique()],
    'Average Price': ['${0:.2f}'.format(purchase_data['Price'].mean())],
    'Number of Purchases': [purchase_data['Price'].count()],
    'Total Revenue': ['${0:.2f}'.format(purchase_data['Price'].sum())],
})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
total_players = purchase_data['SN'].nunique()

male_df = purchase_data.query('Gender == "Male"')
female_df = purchase_data.query('Gender == "Female"')
others_df = purchase_data.query('Gender != "Female" and Gender != "Male"')

male_count = male_df['SN'].nunique()
female_count = female_df['SN'].nunique()
others_count = others_df['SN'].nunique()

male_perc = round(100 * male_count / total_players, 2)
female_perc = round(100 * female_count / total_players, 2)
others_perc = round(100 * others_count / total_players, 2)

df = pd.DataFrame({
    'Total Count': [male_count, female_count, others_count],
    'Percentage of Players': [male_perc, female_perc, others_perc],
})
df.rename({
    0: 'Male',
    1: 'Female',
    2: 'Other / Non-Disclosed'
})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
female_df = purchase_data.query('Gender == "Female"')
male_df = purchase_data.query('Gender == "Male"')
others_df = purchase_data.query('Gender != "Female" and Gender != "Male"')

female_count = female_df['Price'].count()
male_count = male_df['Price'].count()
others_count = others_df['Price'].count()

avg_female_price = '${0:.2f}'.format(female_df['Price'].mean())
avg_male_price = '${0:.2f}'.format(male_df['Price'].mean())
avg_others_price = '${0:.2f}'.format(others_df['Price'].mean())

total_female_price = '${0:.2f}'.format(female_df['Price'].sum())
total_male_price = '${0:.2f}'.format(male_df['Price'].sum())
total_others_price = '${0:.2f}'.format(others_df['Price'].sum())

avg_female_per_person = '${0:.2f}'.format(female_df['Price'].sum() / female_df['SN'].nunique())
avg_male_per_person = '${0:.2f}'.format(male_df['Price'].sum() / male_df['SN'].nunique())
avg_others_per_person = '${0:.2f}'.format(others_df['Price'].sum() / others_df['SN'].nunique())

df = pd.DataFrame({
    'Gender': ['Female', 'Male', 'Other / Non-Disclosed'],
    'Total Count': [female_count, male_count, others_count],
    'Average Purchase Price': [avg_female_price, avg_male_price, avg_others_price],
    'Total Purchase Value': [total_female_price, total_male_price, total_others_price],
    'Avg Total Purchase per Person': [avg_female_per_person, avg_male_per_person, avg_others_per_person]
})
df.set_index('Gender')
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top5 = pd.DataFrame(purchase_data.groupby('SN')['Price'].sum()).sort_values('Price', ascending=False).head(5).index.values

purchase_count = []
avg_purchase_price = []
total_purchase_value = []
for sn in top5:
    sn_df = purchase_data.query('SN == "{}"'.format(sn))
    purchase_count.append(sn_df['Price'].count())
    avg_purchase_price.append('${0:.2f}'.format(sn_df['Price'].mean()))
    total_purchase_value.append('${0:.2f}'.format(sn_df['Price'].sum()))

df = pd.DataFrame({
    'SN': top5,
    'Purchase Count': purchase_count,
    'Average Purchase Price': avg_purchase_price,
    'Total Purchase Value': total_purchase_value,
})
df.set_index('SN')
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top5 = pd.DataFrame(purchase_data.groupby('Item ID')['Price'].count()).sort_values('Price', ascending=False).head(5).index.values

purchase_count = []
avg_purchase_price = []
total_purchase_value = []
item_names = []
for item_id in top5:
    item_df = purchase_data[purchase_data['Item ID'] == item_id]
    item_names.append(item_df['Item Name'].iloc[0])
    purchase_count.append(item_df['Price'].count())
    avg_purchase_price.append('${0:.2f}'.format(item_df['Price'].mean()))
    total_purchase_value.append('${0:.2f}'.format(item_df['Price'].sum()))

df = pd.DataFrame({
    'Item ID': top5,
    'Item Name': item_names,
    'Purchase Count': purchase_count,
    'Item Price': avg_purchase_price,
    'Total Purchase Value': total_purchase_value,
})
df.set_index(['Item ID', 'Item Name'])
```
